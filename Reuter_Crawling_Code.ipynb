```python
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
import time, os, shutil, re
import requests
from selenium import webdriver
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# 크롤링 코드

# 페이지를 반복하기.

keywords = ['car','shipbuilding','steel','oil price','semiconductor']

driver = webdriver.Chrome()
# driver.implicitly_wait(10)

for keyword in keywords:
    
    title_ls = []
    category_ls = []
    date_ls = []
    
    try:
        for page in range(0,5000,10):
            driver.get('https://www.reuters.com/site-search/?query={}&offset={}'.format(keyword,page))
            time.sleep(1)
            html = driver.page_source
            soup = BeautifulSoup(html,'html.parser')
            lis = soup.select('#fusion-app > div > div> div > div > div > div > ul > li')

            print('-'*60)
            print(keyword)
            print(page)
            print('-'*60)

            for li in lis:
                try:
                    title_lis = li.select('div > div.media-story-card__body__3tRWy > a')[0].text.split()[:-3]
                    title_lis = ' '.join(title_lis)
                    print(title_lis)
                    title_ls.append(title_lis)      

                    date_lis =li.select('div > div.media-story-card__body__3tRWy > time')[0].text
                    print(date_lis)
                    date_ls.append(date_lis)

                    category_lis = li.select('div > div.media-story-card__body__3tRWy > span > span')[0].text
                    print(category_lis)
                    category_ls.append(category_lis)

                except:
                    title_lis = li.select('div > a')[0].text
                    print(title_lis)
                    title_ls.append(title_lis)

                    date_lis =li.select('div > time')[0].text
                    print(date_lis)
                    date_ls.append(date_lis)

                    category_lis = li.select('div > span > span')[0].text
                    print(category_lis)
                    category_ls.append(category_lis)
        
        df = pd.DataFrame({'title':title_ls,'date':date_ls,'category':category_ls})
        df.to_csv(f'reuters_{keyword}.csv')
        
    except IndexError:
            print('index out of range')
            
```
