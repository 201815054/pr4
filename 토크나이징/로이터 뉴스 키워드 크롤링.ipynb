```python
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
import time, os, shutil, re
import requests
```

```python
# 드라이버 어떤걸 쓸건지
driver = webdriver.Chrome()

#어떤 페이지를 열건지
url = 'https://www.reuters.com/site-search/?query='
driver.get(url)

#첫 페이지 들어가서 서치버튼 클릭
search = driver.find_element_by_css_selector('#searchreuters')

#키워드 입력
search.send_keys('car')

# 검색버튼 누름
driver.find_element_by_css_selector('#fusion-app > div > div.search-layout__body__1FDkI > div.search-layout__main__L267c > div > form > div > button > svg').click()

# 리스팅할 변수 지정 
lis = driver.find_elements_by_css_selector('#fusion-app > div > div.search-layout__body__1FDkI > div.search-layout__main__L267c > div > div > div.search-results__sectionContainer__34n_c > ul >li')

# 타이틀, 데이트, 카테고리를 저장할 리스트 지정 
title_lis = []
category_lis = []
date_lis = []
    
#검색해서 나온창의 타이틀, date, 카테고리경로를 리스팅경로 변수에서 하나씩 읽어옴
for li in lis:
    # 사진이 있는 타이틀,데이트,카테고리를 리스트에 저장
        try:
            title = li.find_element_by_css_selector('div > div > a').text#.split('\n')[0]
            print(title)
            title_lis.append(title)
            category = li.find_element_by_css_selector('div > div > span > span').text
            print(category)
            category_lis.append(category)
            date = li.find_element_by_css_selector('div > div > time').text
            print(date)
            date_lis.append(date)
    # 사진이 없으면 바로 except 구문으로 넘어감 
        except:
            title = li.find_element_by_css_selector('div > a ').text
            print(title)
            title_lis.append(title)
            category = li.find_element_by_css_selector('div > span > span').text
            print(category)
            category_lis.append(category)
            date = li.find_element_by_css_selector('div > time').text
            print(date)
            date_lis.append(date)

# 다음페이지로 넘어가기 
driver.find_element_by_css_selector('#fusion-app > div > div.search-layout__body__1FDkI > div.search-layout__main__L267c > div > div > div.search-results__pagination__2h60k > button:nth-child(3) > div').click()

```

```python
# 크롤링 코드

# 페이지를 반복하기.
keywords = ['car','advertisement','shipbuilding','steel','oil price','semiconductor']
title_ls = []
category_ls = []
date_ls = []
driver.implicitly_wait(10)

for keyword in keywords:
    for page in range(0,5000,10):
        driver.get('https://www.reuters.com/site-search/?query={}&offset={}'.format(keyword,page))
        
        html = driver.page_source
        soup = BeautifulSoup(html,'html.parser')
        lis = soup.select('#fusion-app > div > div> div > div > div > div > ul > li')

        print('-'*60)
        print(keyword)
        print(page)
        print('-'*60)
        
        for li in lis:
            try:
                title_lis = li.select('div > div.media-story-card__body__3tRWy > a')[0].text.split()[:-3]
                title_lis = ' '.join(title_lis)
                print(title_lis)
                title_ls.append(title_lis)      
                
                date_lis =li.select('div > div.media-story-card__body__3tRWy > time')[0].text
                print(date_lis)
                date_ls.append(date_lis)
                
                category_lis = li.select('div > div.media-story-card__body__3tRWy > span > span')[0].text
                print(category_lis)
                category_ls.append(category_lis)
                
            except:
                title_lis = li.select('div > a')[0].text
                print(title_lis)
                title_ls.append(title_lis)
                
                date_lis =li.select('div > time')[0].text
                print(date_lis)
                date_ls.append(date_lis)
                
                category_lis = li.select('div > span > span')[0].text
                print(category_lis)
                category_ls.append(category_lis)
        
    pd.DataFrame({'title':title_ls,'date':date_ls,'category':category_ls}).to_csv(f'reuters_{keyword}.csv')
    
                
                
```
